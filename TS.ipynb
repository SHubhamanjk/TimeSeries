```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sms
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv('sales.csv')
data
```

```python
data['Date']=pd.to_datetime(data['Date'])
data.set_index('Date',inplace=True)
data
```

```python
data['Amount'].plot()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adf_test(d):
    result=adfuller(d)
    print('P value ->',result[1])
    if result[1] <= 0.05:
        print("Reject the null hypothesis. Data is stationary")
    else:
        print("Accept the null hypothesis, Data is not stationary ")
```

```python
adf_test(data['Amount'])
```

```python
data['First Difference']=data['Amount']-data['Amount'].shift(1)
```

```python
adf_test(data['First Difference'].dropna())
```

```python
data['Second Difference']=data['First Difference']-data['First Difference'].shift(1)
```

```python
adf_test(data['Second Difference'].dropna())
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
plot1 = plot_acf(data["Second Difference"].dropna())  # this is basically q
```

```python
plot2 = plot_pacf(data["Second Difference"].dropna()) # this is p
```

```python
data['Seasonal Difference']=data['Amount']-data['Amount'].shift(12)    # this is used in sarima
```

```python
data
```

```python
plot3 = plot_acf(data["Seasonal Difference"].dropna())
```

```python
plot4 = plot_pacf(data["Seasonal Difference"].dropna())
```

```python
from datetime import datetime
```

```python
train_data=data
```

```python
from statsmodels.tsa.arima.model import ARIMA
modela=ARIMA(train_data['Amount'],order=(17,2,3)) #order=(p , d , q)
arima_modal = modela.fit()
```

```python
predicteda = arima_modal.predict(start=datetime(2003,1,1),end=datetime(2008,12,1))
train_data['Predicted_arima']=predicteda
```

```python
train_data
```

```python
train_data[['Amount','Predicted_arima']].plot(figsize=(10,5))
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
smodel=SARIMAX(train_data['Amount'],order=(2,0,4),seasonal_order=(0,1,0,12))  #order=(p , d , q)  # seasonal diffrence = (p,d,q,range)
model=smodel.fit()
```

```python
predicted = model.predict(start=datetime(2003,1,1),end=datetime(2008,12,1))
train_data['Predicted']=predicted
```

```python
train_data
```

```python
train_data[['Amount','Predicted']].plot(figsize=(10,5))
```

```python
predictedarima = arima_modal.predict(start=datetime(2009,1,1),end=datetime(2009,12,1))
predictedsarima = model.predict(start=datetime(2009,1,1),end=datetime(2009,12,1))
train_data['Predictedarima']=predictedarima
train_data['Predictedsarima']=predictedsarima
# train_data[['Amount']].plot(xlim=['2005-01-01','2009-01-01'],figsize=(10,5))
# predicted1
```

Basically with the help of all these we predict the future trend on non stationary data  with the help of arima and sarima ....

Steps included in :

First of all we convert the non stationay data to stationary data using dickkey fully test and shift ...

then no of diffrences till we got stationary data is assigned to d

and we use pacf to find p and acf to q

after finding we can train the arima model using order (p,d,q) that's it for arima


in sarima we take seasonal diffrence and use pacf to find p and acf to q then we provide parameters order in same form as arima and another one that seasonal_order in this(0,d,0,range) form...

that's it.........
